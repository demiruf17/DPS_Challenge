### Visualization for the number of accidents per category

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from data_model import DataModel
```

```python
dm = DataModel("data/220511_monatszahlenmonatszahlen2204_verkehrsunfaelle.csv")
df = dm.df[dm.df["JAHR"] < 2021].reset_index(drop=True)


ax = df[["JAHR", "MONATSZAHL", "WERT"]].groupby(["JAHR", "MONATSZAHL"]).sum().unstack()\
    .plot(kind = "bar", figsize=(18, 10), title="Number of Accidents per Category")

ax.set_xlabel("JAHR")
ax.set_ylabel("WERT")
ax.grid()
```

```python
ax = df[["JAHR", "AUSPRAEGUNG", "WERT"]].groupby(["JAHR", "AUSPRAEGUNG"]).sum().unstack()\
    .plot(kind = "bar", figsize=(18, 10), title="Number of Accidents per Type")

ax.set_xlabel("JAHR")
ax.set_ylabel("WERT")
ax.grid()
```
